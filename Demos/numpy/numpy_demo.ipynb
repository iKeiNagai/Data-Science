## LaTeX

### Inline Equation
You can include an inline equation by wrapping the LaTeX code with single `$` symbols.

For example: `Euler's identity: $e^{i\pi} + 1 = 0$` will display as:

Euler's identity: $e^{i\pi} + 1 = 0$

### Block Equation
To create a block equation, wrap the LaTeX code with double `$$` symbols.

For example: `$$\int_{a}^{b} f(x)dx = F(b) - F(a)$$`
will display as:
$$\int_{a}^{b} f(x) \, dx = F(b) - F(a)$$

## Determinant and Rank

````python
A = np.array([[1, 0, 1], 
              [0, 4, 5], 
              [0, 0, 2]])
A
````

## Eigenvectors and Eigenvalues

````python
A = np.array([[5, 0, 0], 
              [0, 2, 2], 
              [0, 2, 2]])
print(A)
````

````python
eigenvalues, eigenvectors = np.linalg.eig(A)
print(eigenvalues) 
print(eigenvectors)
````

##  Solving Linear Systems

````python
A = np.array([[1, 2, 5], 
              [2, 4, 1], 
              [1, 0, 2]])
b = np.array([1, 2, 0])

# Solving Ax = b
x = np.linalg.solve(A, b)
x
````

## Singular Value Decomposition (SVD)

````python
A = np.array([[1, 2], [3, 4]])

# Singular Value Decomposition
U, Sigma, V = np.linalg.svd(A)

print(U)
print(Sigma)
print(V)
````

````python
Sigma_matrix = np.diag(Sigma)
Sigma_matrix
````

````python
import matplotlib.pyplot as plt
the_cat = plt.imread('./cat_image.jpg')
the_cat.shape
````

````python
U, Sigma, V = np.linalg.svd(the_cat)

reduced_images = []
sizes = [1, 2, 3, 4, 5, 10, 50, 100, 3308]
for i in sizes:
    # We choose i first rows of U
    U_reduced = U[:, :i]; 
    
    # We convert i first elements of Sigma_vector into a diagonal matrix Sigma_matrix_reduced
    Sigma_reduced = np.diag(Sigma[:i])
    
    # We choose i first columns of V
    V_reduced = V[:i, :];
    
    reduced_images.append(U_reduced @ Sigma_reduced @ V_reduced)  
````

````python
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(20,15))
i = 0
for row in ax:
    for col in row:
        current_image = reduced_images[i]
        col.imshow(current_image, cmap='gray', vmin=0, vmax=255)
        col.set_title("# of components = " + str(sizes[i]))
        col.axis('off')
        i += 1
        
plt.savefig("reduced_image")        
plt.show()
````
